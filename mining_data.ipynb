```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris

# Load dataset (contoh menggunakan dataset iris)
data = load_iris()
X = data.data
y = data.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create PCA object with 1 principal component
pca = PCA(n_components=1)

# Fit PCA on the training data and transform it
X_train_pca = pca.fit_transform(X_train)

# Transform the testing data using the same PCA model
X_test_pca = pca.transform(X_test)

# Get the explained variance ratio
explained_variance = pca.explained_variance_ratio_

# Print the explained variance
print("Explained variance ratio:", explained_variance)

# Print transformed data to verify
print("Transformed X_train:", X_train_pca)
print("Transformed X_test:", X_test_pca)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris

# Load dataset (contoh menggunakan dataset iris)
data = load_iris()
X = data.data

# List to hold inertia values
inertias = []

# Loop over different numbers of clusters
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Plot the inertia values
plt.plot(range(1, 11), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import pandas as pd

# Load dataset
data = load_iris()
X = data.data
y = data.target

# Reduce dimensions to 2 using PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Split the data into training and testing sets
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state=42)

# Fit KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_train_pca)

# Predict cluster labels for the training data
y_train_clusters = kmeans.labels_

# Create a DataFrame with the original labels and cluster labels
df = pd.DataFrame(X_train_pca, columns=['PCA1', 'PCA2'])
df['True_Label'] = y_train
df['Cluster_Label'] = y_train_clusters

# Print the first few rows of the dataset with cluster labels
print(df.head())

# Plot the clusters
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title('KMeans Clustering with 3 Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.metrics import homogeneity_score, completeness_score, v_measure_score, adjusted_rand_score, adjusted_mutual_info_score, silhouette_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Load dataset
data = load_iris()
X = data.data
y = data.target

# Reduce dimensions to 2 using PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Split the data into training and testing sets
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state=42)

# Fit KMeans clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_train_pca)

# Predict the clusters for the training set
y_pred = kmeans.labels_

# Plot the clusters
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_pred, cmap='viridis')
plt.title('KMeans Clustering with 2 Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# Evaluate clustering performance
print(f"Homogeneity: {homogeneity_score(y_train, y_pred):.3f}")
print(f"Completeness: {completeness_score(y_train, y_pred):.3f}")
print(f"V-measure: {v_measure_score(y_train, y_pred):.3f}")
print(f"Adjusted Rand Index: {adjusted_rand_score(y_train, y_pred):.3f}")
print(f"Adjusted Mutual Information: {adjusted_mutual_info_score(y_train, y_pred):.3f}")
print(f"Silhouette Score: {silhouette_score(X_train_pca, y_pred):.3f}")

```

```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Load dataset
data = load_iris()
X = data.data
y = data.target

# Reduce dimensions to 2 using PCA for visualization
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Split the data into training and testing sets
X_train_pca, X_test_pca, y_train, y_test = train_test_split(X_pca, y, test_size=0.3, random_state=42)

# Fit KMeans clustering
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans.fit(X_train_pca)

# Predict the clusters for the training set
y_pred = kmeans.labels_

# Plot the clusters
plt.scatter(X_train_pca[:, 0], X_train_pca[:, 1], c=y_pred, cmap='viridis')
plt.title('KMeans Clustering with 2 Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()

# Count the number of points in each cluster
cluster_counts = np.bincount(y_pred)
print("Cluster counts:", cluster_counts)

# Find the cluster with the most points
most_common_cluster = np.argmax(cluster_counts)
print(f"The cluster with the most points is cluster {most_common_cluster} with {cluster_counts[most_common_cluster]} points.")
```
